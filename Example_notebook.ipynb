## Import libraries

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mardia import Mardia
```

## Import data

More details in the report

```python
data = pd.read_csv('decathlon.txt', delimiter=' ', index_col=0)
total_points = data['Points']
data = data.drop('Points', axis=1)
data
```

## Obtain the first and second most skewed projections

```python
X = np.array(data)
exp = Mardia(data)
c_s, f_vals = exp.run(k=2, vis=False)
```

```python
fig, ax = plt.subplots()
sp_1 = exp.Z.dot(c_s[0])
sp_2 = exp.Z.dot(c_s[1])
ax.scatter(sp_1, sp_2)

for i, txt in enumerate(data.index):
    ax.annotate(txt, (sp_1[i], sp_2[i]))

ax.set_xlabel('First skewed projection', fontsize=14)
ax.set_ylabel('Second skewed projection', fontsize=14)
plt.show()
```
